### Data Preparation for Llama Fine-tuning using Anyscale

```python
import pandas as pd
import json
```

```python
df = pd.read_excel("linkedin_posts.xlsx")
```

```python
df.head(5)
```

```python
def row_to_json(row):
    return {
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": row['prompt']},
            {"role": "assistant", "content": row['completion']}
        ]
    }
```

```python
df['text'] = df.apply(row_to_json, axis=1)
```

```python
df.head(5)
```

```python
df['text'][0]
```

```python
print(df['text'][0])
```

```python
with open('linkedin_posts_dataset.jsonl', 'w') as f:
    for json_line in df['text']:
        f.write(json.dumps(json_line) + '\n')
```
